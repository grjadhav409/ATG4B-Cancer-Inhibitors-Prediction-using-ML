```python
smiles1 = df['canonical_smiles'].tolist()
```

```python

#featurizer = dc.feat.CircularFingerprint(size=1024)
featurizer = dc.feat.MACCSKeysFingerprint()
#featurizer=dc.feat.ConvMolFeaturizer(per_atom_fragmentation=False)
#featurizer = dc.feat.PubChemFingerprint() #required rdkit and pubchempy installed
#featurizer = dc.feat.RDKitDescriptors()
#featurizer = dc.feat.MordredDescriptors(ignore_3D=True)



discriptors = featurizer.featurize(smiles1)
```

```python
X_MACCSKeysFingerprint= pd.DataFrame(data=discriptors)
X_MACCSKeysFingerprint
```
